```python
#loading some stuff 
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import pickle
```

```python
#now loading the pickle file containing the ADL data
with open(base_path + "LatestADL.pkl", "rb") as input_file:
    ADL = pickle.load(input_file)
print("Read ADL from pickle file")

#now loading the pickle file containing the returns (can be updated)
with open(base_path + "AllReturns.pkl", "rb") as input_file:
    RET = pickle.load(input_file)
print("Read return data from pickle file")

#and also the RFT containing the latest RFT settings
with open(base_path + "LatestRFTSetting.pkl", "rb") as input_file:
    RFT = pickle.load(input_file)
print("Read RFT from pickle file")
```

```python
#got it that you need iteration to manipulate creation date from April 2019
OK_hi_ids = []
cnt = 0
for hi_id, adl_hi in ADL.groupby('ADLRawData_HearingInstrumentId'):
    cnt = cnt + 1
    if (cnt % 1000 == 0):
        print('HI = ',cnt)
    
    #at least 10 ADL packages
    count_in_period = adl_hi[(adl_hi['ADLRawData_CreationTime']>'2019-04-01') & (adl_hi['ADLRawData_CreationTime'].max() > '2019-08-01') \
                             & (adl_hi['ADLRawData_CreationTime'].min()<'2019-02-01')].shape[0]

    if count_in_period > 10:
        OK_hi_ids.append(hi_id)
print("Found " + str(len(OK_hi_ids)) + " HI's with more than 10 ADL packages in period")

```

```python
# Make subselection
sub_ADL = ADL[ADL['ADLRawData_HearingInstrumentId'].isin(OK_hi_ids)]
# Reset the index
sub_ADL = sub_ADL.set_index(np.arange(sub_ADL.shape[0]))
sub_ADL.rename(columns={'ADLRawData_HearingInstrumentId':'HearingInstrumentId'}, inplace = True)
```

```python
#now prepare RFT to join the returns to the ADL time series 
RFT_Setting = RFT[['SerialNumber','HearingInstrumentId']]
RET_basis = RET[['SerialNumber','RootCause', 'ReturnDate']]
```

```python
#now try to merge the dataframes 
tmp = RFT_Setting.merge(RET_basis, on='SerialNumber', how='inner')
use_time_returns = pd.merge(sub_ADL,tmp.reset_index(), on ='HearingInstrumentId', how='inner')
```

```python
cleaned_df = []
cnt = 0
hi_id_return = []
for hi_id, group in use_time_returns.groupby('HearingInstrumentId'):
    if(cnt%1000 == 0):
        print(cnt)
    cleaned_df.append(group.drop_duplicates(subset='ADLRawData_CreationTime'))
    hi_id_return.append(hi_id)
    cnt += 1
    
cleaned_df = pd.concat(cleaned_df)    
cleaned_df.set_index(cleaned_df['ADLRawData_CreationTime'],inplace=True)
sub_ADL.set_index(sub_ADL['ADLRawData_CreationTime'],inplace=True)
```

### Now filter the return cause

```python
#•	Wax
#•	Cera
#•	Causa: cerume
#•	귀지
#•	Voks
#•	Cérumen
#•	耳垢詰まり
#•	Cerumen

cause = ['Wax','Cera','Causa: cerume','귀지','Voks','Cérumen','耳垢詰まり','Cerumen'] #here list of all expressions describing wax 

doFilterCauses = 1
if doFilterCauses: 
    #filter out root cause wax         
    regstr = '|'.join(cause)
    cleaned_df = cleaned_df[cleaned_df['RootCause'].str.contains(regstr)]
    print('Changed the cause to', cause)

print('Number of elements with cause',cleaned_df.shape[0])
```

```python
#calculating time between last ADL and return
diff_days = []
hi_id_last_ADL = []
cutoff = 30#days

cnt = 0
for hi_id, group in cleaned_df.groupby('HearingInstrumentId'):        
        if(cnt%1000 == 0):
            print(cnt)            
        
        cnt += 1
        
        re_date = group['ReturnDate'].iloc[0]
        
        if (group.iloc[group.index < re_date].empty):
            continue
        
        #get rid of the timezone
        diff_days.append((pd.Timestamp(re_date) - \
        pd.Timestamp(group.iloc[group.index < re_date].index[-1][:-6])).days)            
        
        #cut-off at 30 days before return
        if (diff_days[-1] > cutoff):
            continue
                
        hi_id_last_ADL.append(hi_id)      
        
```

```python
#create new field VolumeControl
sub_ADL['VolumeControl']=(sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_0'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_1'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_2'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_3'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_4'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_5'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_6'] +
sub_ADL['ADLProperties_MeanVolumeControlLevelPerPreset_7'])/ 8
```

```python
#now do the filtering of data before return
#returns: time series before returns
#non-returns: sequence of the same length (12 weeks) with random starting time
from random import randrange

#the relevant fields are the following 
new_keys = ['ADLProperties_TotalUseTime','ADLProperties_NumberOfDeviceBoots',\
        'ADLProperties_NumberOfPowerDowns','ADLProperties_NumberOfPresetSwitches',\
        'VolumeControl','HearingInstrumentId']

sub_sub_ADL = sub_ADL[new_keys]

cnt = 0
prereturn = []

for hi_id, group in sub_sub_ADL.groupby('HearingInstrumentId'):        
        if(cnt%1000 == 0):
            print(cnt)
            
        if (hi_id in hi_id_last_ADL):  
            
            tmp = cleaned_df[cleaned_df['HearingInstrumentId'] == hi_id]
            re_date = tmp['ReturnDate'].iloc[0]            
                                       
            idx = group.index < re_date
            
            time_series = group[idx]                         
            
            prereturn.append(time_series)            
          
        else:
                        
            time_series = group #.iloc[0:randrange(group.shape[0])]                         
            
            prereturn.append(time_series) 
                     
        cnt += 1        

for_resample_df =pd.concat(prereturn)
```

```python
#now resample all columns for further analysis
resampled_time_series= []
appended_df = []

cnt = 0
appended_df = []
for hi_id, group in for_resample_df.groupby('HearingInstrumentId'):
    cnt += 1
    if (cnt % 1000 == 0):
        print("Resampled " + str(cnt) + " ADLs")
    time_series = group    
    time_series = time_series.set_index(pd.to_datetime(group.index))
    time_series = time_series.loc[~time_series.index.duplicated(keep='last')]
    time_series = time_series.sort_index()

    r_time_series = time_series[new_keys[:-1]].diff().resample('W').sum().\
    interpolate(method = 'linear') 
        
    for colname, col in r_time_series.iteritems():
        bin = col ==0
        if bin.iloc[-1]==True:
            continue
        for i in np.arange(col.shape[0]):
            start_row = i
            replace_rows = []
            n = 0
            #print("Start row : " + str(start_row))
            while (bin[start_row] == True):
                replace_rows.append(start_row)
                start_row = start_row + 1
                n = n + 1
                #print("Replace rows : " + str(replace_rows))
            if len(replace_rows) > 0:
                bin.iloc[replace_rows] = False            
                replace_value = col.iloc[start_row] / (n+1)
                col.iloc[replace_rows] = replace_value
                col.iloc[start_row] = replace_value

    resampled_time_series = r_time_series / 7
    resampled_time_series['ADLRawData_HearingInstrumentId'] = hi_id
         
    appended_df.append(resampled_time_series)

new_resampled_df = pd.concat(appended_df)
```

```python
#create class labels for classification
new_resampled_df['class'] = 0
new_resampled_df.loc[new_resampled_df['ADLRawData_HearingInstrumentId'].isin(hi_id_last_ADL), 'class'] = 1
```

```python
#now create category of isVolumenChange; count volume change per week >= 1 for each individual
new_resampled_df['isVolumeChange'] = 0
new_resampled_df.loc[new_resampled_df['VolumeControl'].diff() != 0,'isVolumeChange'] = 1
```

```python
#prepare for plotting the time course of the five predictors for X weeks prior to last ADL bedore 
#return

plot_data_1 = []
plot_data_2 = []
plot_data_3 = []
plot_data_4 = []
plot_data_5 = []

class_ = []
cnt = 0 
nrWeeks = 12
#now prepare the numpy for plotting the data 
for hi_id, group in new_resampled_df.groupby('ADLRawData_HearingInstrumentId'):
    if (cnt%1000 == 0):
        print(cnt)
    
    if (hi_id in hi_id_last_ADL):
    
        tmp1 = group['ADLProperties_TotalUseTime'].iloc[-nrWeeks:].values     
        tmp2 = group['ADLProperties_NumberOfPresetSwitches'].iloc[-nrWeeks:].values    
        tmp3 = group['ADLProperties_NumberOfPowerDowns'].iloc[-nrWeeks:].values
        tmp4 = group['isVolumeChange'].iloc[-nrWeeks:].values
        tmp5 = group['ADLProperties_NumberOfDeviceBoots'].iloc[-nrWeeks:].values
        
    else:
        rnd = randrange(group.shape[0] - nrWeeks)
        tmp1 = group['ADLProperties_TotalUseTime'].iloc[rnd:rnd + nrWeeks].values     
        tmp2 = group['ADLProperties_NumberOfPresetSwitches'].iloc[rnd:rnd + nrWeeks].values    
        tmp3 = group['ADLProperties_NumberOfPowerDowns'].iloc[rnd:rnd + nrWeeks].values
        tmp4 = group['isVolumeChange'].iloc[rnd:rnd + nrWeeks].values
        tmp5 = group['ADLProperties_NumberOfDeviceBoots'].iloc[rnd:rnd + nrWeeks].values
    
    #use time 
    if (tmp1.shape[0] != 12):
        continue
        
    if(np.any(tmp1 < 0)):
        continue
                
    if(np.any(tmp1 > 24)):
        continue
    
    #presets
    if (tmp2.shape[0] != 12):
        continue
        
    if(np.any(tmp2 < 0)):
        continue
                
    if(np.any(tmp2 > 1000)):
        continue
    
    #power downs 
    if (tmp3.shape[0] != 12):
        continue
        
    if(np.any(tmp3 < 0)):
        continue
                
    if(np.any(tmp3 > 1000)):
        continue
    
    #volume change 
    if (tmp4.shape[0] != 12):
        continue
        
    if(np.any(tmp4 < -100)):
        continue
                
    if(np.any(tmp4 > 100)):
        continue
    
    #re-boots 
    if (tmp5.shape[0] != 12):
        continue
        
    if(np.any(tmp5 < 0)):
        continue
                
    if(np.any(tmp5 > 1000)):
        continue
    
    plot_data_1.append(tmp1)
    plot_data_2.append(tmp2)
    plot_data_3.append(tmp3)
    plot_data_4.append(tmp4)
    plot_data_5.append(tmp5)   
    
    class_.append(group['class'].iloc[0])
    cnt += 1    
    
plot_data_1 = np.array(plot_data_1)
plot_data_2 = np.array(plot_data_2)
plot_data_3 = np.array(plot_data_3)
plot_data_4 = np.array(plot_data_4)
plot_data_5 = np.array(plot_data_5)
class_labels_array = np.array(class_)
```

```python
#now try to do the logistic regression for return prediction (prepare data)
X_CNN = np.array([plot_data_1.flatten(), 
plot_data_2.flatten(),
plot_data_3.flatten(),
plot_data_4.flatten(),
plot_data_5.flatten()]).transpose()

Y_ = class_labels_array
data_names = np.array(['Use_time_%d' % i for i in range(11,-1,-1)] + \
                      ['Presets_%d' % i for i in range(11,-1,-1)] + \
['Power_downs_%d' % i for i in range(11,-1,-1)] + \
                      #['Volume_changes_%d' % i for i in range(11,-1,-1)] + \
['re-boots_%d' % i for i in range(11,-1,-1)])
                     
```

```python
#get the right format for the inputs of CNN
X_CNN = X_CNN.reshape(int(X_CNN.shape[0] / 12), 12, 5)
Y_CNN = np.zeros((Y_.shape[0],1))
Y_CNN[np.where(Y_ == 1)] = 1
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv1D
from keras.layers import MaxPooling1D
from keras.layers import Flatten
from keras.layers import Dropout
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras import models
```

```python
#now one-hot encode
Y_OHC = to_categorical(Y_CNN)
```

```python
#make train, test data
trainX, testX, trainY, testY = train_test_split(X_CNN, Y_OHC, test_size=0.20, random_state=42)
n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], trainY.shape[1]
```

```python
#try an easy 1D conv 
model = Sequential()
model.add(Conv1D(16,4, activation = 'relu', input_shape=(n_timesteps,n_features)))
model.add(Dropout(0.5)) 
model.add(MaxPooling1D(pool_size = 2))
model.add(Flatten())
model.add(Dense(n_outputs, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
        
```

```python
#model fitting
model.fit(trainX, trainY, epochs=10, batch_size=32, validation_split = 0.2)
```

```python
#confusion matrix 
from sklearn.metrics import confusion_matrix
confusion_matrix(testY.argmax(axis=1), model.predict_classes(testX))
```

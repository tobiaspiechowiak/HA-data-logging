```python
hi_id = []
target_freq = [250, 500, 1000, 2000, 4000, 6000, 8000]
interp_threshold = []
interp_threshold_dict = {}

#looping over all hi_ids and extracting audiogram 
for idx in range(0,AUD['HearingInstrumentId'].shape[0]):
    aud_freq =  []
    threshold = []
    if (idx%10000 == 0):
        print('HI =',idx)
    if (len(AUD['PayLoadContent'][idx]['AirConduction']['CurvePoints']) == 0):
        print('Empty array')
        continue
    
    for freq_idx in range(0,len(AUD['PayLoadContent'][idx]['AirConduction']['CurvePoints'])):
                
        
        aud_freq.append(AUD['PayLoadContent'][idx]\
                        ['AirConduction']['CurvePoints'][freq_idx]['Frequency'])
        
        threshold.append(AUD['PayLoadContent'][idx]\
                        ['AirConduction']['CurvePoints'][freq_idx]['Intensity'])
    
    
    interp_threshold.append(np.interp(target_freq, aud_freq, threshold))        
    hi_id.append(AUD['HearingInstrumentId'][idx])  
    
    interp_threshold_dict[AUD['HearingInstrumentId'][idx]] = interp_threshold[-1]

```

```python
#classification of standard audiograms
def aud_classify(threshold):    
    std_aud = 10*[None]
    labels = ['N1','N2','N3','N4','N5','N6','N7','S1','S2','S3']
    import numpy as np
    #Flat and moderately sloping 
    std_aud[0] = [10, 10, 10, 15, 30, 40] #N1 
    std_aud[1] = [20, 20, 25, 35, 45, 50] #N2
    std_aud[2] = [35, 35, 40, 50, 60, 65] #N3
    std_aud[3] = [55, 55, 55, 65, 75, 80] #N4
    std_aud[4] = [65, 70, 75, 80, 80, 80] #N5
    std_aud[5] = [75, 80, 85, 90, 100, 100] #N6
    std_aud[6] = [90, 95, 105, 105, 105, 105] #N7
    #Steep slooping 
    std_aud[7] = [10, 10, 10, 15, 55, 70] #S1
    std_aud[8] = [20, 20, 25, 55, 95, 95] #S2
    std_aud[9] = [30, 35, 60, 75, 80, 85] #S3
    
    return(labels[np.argmin(((np.array(std_aud) - threshold)**2).mean(axis=1)**(1/2))])
    
```

```python
aud_hi = aud_hi.drop_duplicates(subset = 'HearingInstrumentId')
```

```python
#classify audiogram into standard classes
aud_classes = {}
cnt = 0
for hi_id, grp in aud_hi.groupby('HearingInstrumentId'):    
    if(cnt%1000 == 0):
        print('Hi =',cnt)
    aud_classes[hi_id] = aud_classify(grp[['250Hz','500Hz','1000Hz','2000Hz','4000Hz','6000Hz']].values)
    cnt = cnt + 1 
```
